## first approach is finding the mean of sw column and filling with the mean value

```python
df.sw.fillna(df.sw.mean(),inplace=True)
df.pl.fillna(df.pl.mean(),inplace=True)# i have fill sw and pl column NaN value by mean value of that column
df
```

## how to fill NaN value by mean of same flower type(v)

```python
a=df[df.v=='Setosa']
df.sw.fillna(a.sw.mean(),inplace=True)
df.pl.fillna(a.pl.mean(),inplace=True)
df
```

## how to deal with string value

```python
df["gender"]='male'
df.iloc[0:4,5]='Female'
df
```

```python
## how to change male and female by respective number

def f(s):
    if s=='male':
        return 0
    else:
        return 1
    
df['sex']=df.gender.apply(f)
df.head()
```
