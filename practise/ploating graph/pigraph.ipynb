```python
import matplotlib.pyplot as plt
```

```python
size=[4,3,5]
plt.pie(size)
plt.axis("equal") # to make pie round
plt.show()
```

```python
plt.title("pie")
label=['A','B','C']
size=[4,3,5]
color=["blue","green","purple"]
plt.pie(size,colors=color,labels=label,autopct='%.1f%%')
plt.axis("equal") # to make pie round
plt.show()
```

```python
plt.title("pie")
label=['A','B','C']
size=[4,3,5]
explode=[0.05,0.05,0]
color=["blue","green","purple"]
plt.pie(size,colors=color,labels=label,autopct='%.1f%%',explode=explode,counterclock=False) # explode means highlighting the one piece
plt.axis("equal") # to make pie round     # counterclock=False means while exploding it happen clock wise default id counterclock wise
plt.show()
```

```python
plt.title("pie")
label=['A','B','C']
size=[4,3,5]
explode=[0.05,0.05,0]
color=["blue","green","purple"]
plt.pie(size,colors=color,labels=label,autopct='%.1f%%',explode=explode,counterclock=False,startangle=90)  
plt.axis("equal")    # start angle means clockwise and start from 100 deg
```
