```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a=sns.load_dataset("flights") #check the data on github  https://github.com/mwaskom/seaborn-data
a
```

```python
sns.relplot(x="passengers",y="month",data=a) # we are specifing data present in a
```

```python
sns.relplot(x="year",y="month",kind="line",data=a)
```

```python
sns.relplot(x="passengers",y="month",hue="month",data=a) # we can visulize the data according to month or year
```

```python
sns.relplot(x="passengers",y="month",hue="year",data=a)
```

```python
b=sns.load_dataset("tips")
b
```

```python
sns.relplot(x="day",y="tip",data=b,kind="line") # kind help to change scatter to line graph
```

# categorical data ploting

```python
sns.catplot(x="day",y="tip",data=b)
```

```python
sns.catplot(x="day",y="tip",kind="violin",data=b) #violin make a graph in violin way
```

```python
sns.catplot(x="day",y="tip",kind="boxen",data=b)
```

```python
sns.boxplot(x="day",y="tip",data=b)
```

```python
sns.boxplot(y="total_bill" , data=b); # indivisual graph
plt.show()
```

```python
sns.pairplot(b,hue = 'day') # A pairplot plot a pairwise relationships in a dataset.
```
