```python
import numpy as np 
x=np.arange(0,4,0.1) 
y=x**3
# plt.plot(x,y, color='black',marker="2")
plt.plot(x,y, color='black',linewidth=0.1) 
plt.plot(x,y,marker='2')
plt.ylabel("x^3")
plt.xlabel("x")
plt.title("matplotlip demo")
plt.show()
```

```python
import numpy as np 
x=np.arange(0,4,0.1) 
y1=x**3
y2=x**2
plt.plot(x,y1, color='black',label="x**3") # label give the info which line belong to whom
plt.plot(x,y2,color="green",label="x**2")


plt.ylabel("y")
plt.xlabel("x")

#plt.axis([0,50,0,100]) # first 1 pair for x axis and another  pair for y axis
plt.legend() # to run the label on line graph

plt.grid() # to make a grid on graph

plt.text(1.5,40,'prasant',fontsize=19) # first 2 number is place where we want to place a text

plt.title("matplotlip demo")
plt.show()
```
