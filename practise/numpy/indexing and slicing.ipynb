```python
import numpy as np
li=np.arange(1,5) # generating array
print(li)
```

```python
li=[1,2,3,4,5]
arr=np.array(li) #converting list to array
print(arr)
```

```python
print(arr.data)
print(arr.shape) # 1d array so shape is 5 (depend upon array)
print(arr.dtype)
print(arr.strides) # it skip 8 byte to store a number
```

```python
list_2d=[[1,2,3],[2,3,4],[6,7,8]]
arr_2d=np.array(list_2d)
print(arr_2d)
```

```python
print(arr_2d.data)
print(arr_2d.shape) 
print(arr_2d.dtype)
print(arr_2d.strides) #24 byte is to skipped to get another row and 8 is value skipped to get another value 
```

slicing for 2d arr

```python
print(list_2d[0][2])
print(arr_2d[0][2])  # slicing is same as list 
print(arr_2d[0,2]) #another method
```

```python
print(list_2d[1][:2]) # row is fix and column is varying
print(arr_2d[1,:2])
```

## indexing and slicing

```python
import numpy as np
li=[1,2,3,4,5]
arr=np.array(li) 
print(arr)
```

```python
print(arr[2]) # we can use indexing in numpy array
print(arr[2:4]) # as well as slicing 
```
